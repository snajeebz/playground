# Reading Data

```python
df=pd.read_csv('/kaggle/input/electronic-card-transactions-dec-2023-new-zealand/electronic-card-transactions-december-2023-csv-tables.csv')
```

# Exploring Data

```python
df['Period']=pd.to_datetime(df['Period'].astype('string'))
```

# Generating Pandas Profile Report

```python
pp.ProfileReport(df)
```

# Visualizations

```python
import plotly.express as px

fig = px.bar(df, x='Period', y="Data_value", color=df.Period.dt.month)
fig.show()
```

```python
fig = px.bar(df['Data_value'][df.UNITS=='Dollars'].groupby(df.Period.dt.year).sum(), title='Total Yearly Transaction Amount')
fig.show()
```

```python
fig = px.pie(df[(df.Period.dt.year==2023) & (df.UNITS=='Dollars')], values='Data_value', names='Period', title='Month wise Transactions in 2023')
fig.show()
```

```python
fig = px.bar(df[(df.Period.dt.year==2023) & (df.UNITS=='Dollars')], y='Data_value', x='Period', color='Series_reference', title='Series wise Monthly Transactions Amount in 2023')
fig.show()
```

```python
fig = px.bar(df['Data_value'][df.UNITS=='Dollars'].groupby(df.Period.dt.month).count(), title='Average Recorded Transactions per month')
fig.show()
```

```python
fig = px.bar(df['Data_value'][df.UNITS=='Percent'].groupby(df.Period.dt.month).count(), title='Average Recorded Transactions by Percent per month')
fig.show()
```

```python


filename = df
target_variable = 'Magnitude'

dft = AV.AutoViz(
    "",
    sep=",",
    depVar=target_variable,
    dfte=df,
    header=0,
    verbose=-1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=20123,
    max_cols_analyzed=20,
    save_plot_dir=None
)
```
